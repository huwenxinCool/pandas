# 区间分组
- qcut()
- cut()
## 什么是数据离散化？
- 连续属性的离散化就是将连续属性的值域上，将值域分成若干个离散的区间，最后用不同的符号或整数值代表不同的区间

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_day = pd.read_csv('stock_day.csv', usecols=['p_change'])
```

```python
p_change = stock_day['p_change']
```

```python
p_change.head()
```

# 自动分组
- qcut(x,q)
- x 数据
- q 分组多少组

```python
qcut = pd.qcut(p_change, 10)
```

```python
# 统计各个区间出现多少次b
qcut.value_counts()
```

## 自定义区间
- pd.cut(x, bins)
- x　要进行离散的数据
- bins 自定义的分组

```python
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
p_counts = pd.cut(p_change, bins)
```

```python
p_counts.value_counts()
```

# one-host 编码
- pandas.get_dummies(data, prefix)
- data: 要进行处理的数据
- prefix: 分组的名字

```python
dummies = pd.get_dummies(qcut, prefix='rise')
```
