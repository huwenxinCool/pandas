```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('stock_day.csv')
```

```python
data.head()
```

# 案列：
    - 股票周几的涨幅大

```python
data.index
```

```python
# 把索引转换成时间序列
date = pd.to_datetime(data.index).weekday
```

```python
date
```

```python
data['week'] = date
```

```python
data.head()
```

```python
# 建立新的一列，判断涨就为１，　不涨就为０
data['posi_neg'] = np.where(data['p_change'] > 0, 1, 0)
```

```python
data.head(10)
```

```python
pd.crosstab?
```

```python
count = pd.crosstab(data['week'], data['posi_neg'])
```

```python
count
```

```python
sums = count.sum(axis=1)
```

```python
sums
```

```python
cross = count.div(sums, axis=0)
cross.plot?
```

```python
cross.plot(kind='bar', figsize=(10, 8))
plt.hlines(0.5, -1, 5, color='r')
plt.show()
```

```python
count.sum(axis=1).astype(np.float32)
```

```python
pro = count.div(count.sum(axis=1).astype(np.float32), axis=0)
```

```python
pro
```

```python
pro.plot(kind='bar', stacked=True)
plt.hlines(0.41, -1, 5, color='r')
plt.vlines(1, 0, 1, color='r')
plt.show()
```
