```python
.0import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_day = pd.read_csv('stock_day/stock_day.csv', usecols=['open', 'high', 'low', 'close', 'p_change'])
```

```python
stock_day.head()
```

```python
# 把字符串的日期转换成pandas的时间类型，为了获取其中的工作日信息
weekdays = pd.to_datetime(stock_day.index).weekday
stock_day['weekday'] = weekdays
stock_day.head()
```

```python
# 生成涨跌幅的好（1）与坏（0）数据
stock_day['positive_negative'] = np.where(stock_day.p_change > 0, 1, 0)
stock_day
```

```python
# 使用交叉表
cross = pd.crosstab(stock_day.weekday, stock_day.positive_negative)
cross
```

```python
# 以行为单位求和
sums = cross.sum(axis=1)
```

```python
cross.div(sums, axis=0)
```

```python
# crosstab有一个简单方法可以计算比例
cross = pd.crosstab(stock_day.weekday, stock_day.positive_negative, normalize='index')
```

```python
cross.plot(kind='bar', figsize=(8, 8))
plt.hlines(0.5, -1, 5, color='r')
plt.show()
```

```python
cross.plot(kind='bar', figsize=(8, 8), stacked=True)
plt.show()
```

## 透视表

```python
pivot = stock_day.pivot_table(index='weekday', columns='positive_negative', values='open', aggfunc='count')
pivot
```

```python
sums = pivot.sum(axis=1)
```

```python
pivot.div(sums, axis=0)
```
