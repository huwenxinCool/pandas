## 画出上海11点到12点的温度变化

```python
import random
```

```python
# 准备数据
# x 是时间，分钟
x = range(60)
# ｙ_shanghai是温度
y_shanghai = [random.uniform(15, 18) for i in x]

# 创建画
plt.figure(figsize=(10, 8), dpi=100)

# 绘制折线图
plt.plot(x, y_shanghai)

# 展示
plt.show()
```

## 加上刻度值　

```python
# 准备数据
# x 是时间，分钟
x = range(60)
# ｙ_shanghai是温度
y_shanghai = [random.uniform(15, 18) for i in x]

# 创建画
plt.figure(figsize=(10, 8), dpi=100)

# 绘制折线图
plt.plot(x, y_shanghai, label="上海", linestyle=":")

# ｘ的刻度
x_ticks_label = ["11点{}分".format(i) for i in range(60)]
# y的刻度
y_ticks = range(40)

# 添加网格 True 是显示网格，　ａｌｐｈａ是透明度
plt.grid(True, alpha=0.1)
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 添加ｘ轴　y轴　的描叙信息以及和标题信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("中午11点到12点温度的变化")

# 多次plot
y_beijing = [random.uniform(1, 3) for i in x]
# 添加第二条折线
plt.plot(x, y_beijing, linestyle="--", color='r', label="北京")
# 显示图列 自动选择区域
plt.legend(loc=0)
# 展示
plt.show()
```

## 画子图

```python
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 创建画
fig, axes = plt.subplots(1, 2, figsize=(10,8), dpi=100)
for  data,p,label, color, h in zip(fig,axes, ["上海", "北京"],["r", "b"], [y_shanghai, y_beijing]):
    # ｘ的刻度
    x_ticks_label = ["11点{}分".format(i) for i in range(60)]
    # y的刻度
    y_ticks = range(40)

    # 添加网格 True 是显示网格，　ａｌｐｈａ是透明度
    p.grid(True, alpha=0.1)
    p.set_xticks(x[::5], x_ticks_label[::5])
    p.set_yticks(y_ticks[::5])

    # 添加ｘ轴　y轴　的描叙信息以及和标题信息
    p.set_xlabel("时间")
    p.set_ylabel("温度")
    p.set_title("中午11点到12点温度的变化")

    # 添加第二条折线
    plt.plot(x, h, linestyle="--", color=color, label=label)
    # 显示图列 自动选择区域
    plt.legend(loc=0)
# 展示
    fig.show()
```
